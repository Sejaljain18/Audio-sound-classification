```python
filename='UrbanSound8K/dog_bark.wav'
```

```python
import IPython.display as ipd
import librosa
import librosa.display
```

```python
#Dog Sound
plt.figure(figsize=(14,5))
data,sample_rate=librosa.load(filename)
librosa.display.waveshow(data,sr=sample_rate)
ipd.Audio(filename)
```
